# Data input (Beef)

```python
AD_eval = False
SL_eval = True
biogas_product = 'None'
RINs_methane = 1.2502995248799291 #USD per kg (liquified)
REC = 0 # Renewable Energy Certificate (USD per MWh)
P_credits = 0 #USD per kgP recovered
CapitalCost_incentive = 0 #% over total capital cost
discount_rate=7

N=100
weighting_type = 'GIS_based'
user_weight=dict()
```

# Import modules

```python
from main_PTechs import main_function
from cereslibrary.IndicatorsDatabase.IndicatorsDatabase import IndicatorsDatabase
from cereslibrary.MCDA.MCDA_module import MCDA_module
```

```python
waste_data          = pd.read_csv('cereslibrary/PTechs/feedstock_data/waste_data_input.csv', sep=",", header=0)
waste_data = waste_data.set_index('Livestock type', drop=False)
waste_data
```

```python
IndianaBeef_df = pd.read_csv('CAFOs_Indiana/CAFOs_Indiana_Beef_Filtered.csv')
IndianaBeef_df = IndianaBeef_df.drop(columns=['Unnamed: 0'])
IndianaBeef_df = IndianaBeef_df.rename(columns={"Physical Address": "Address"})
IndianaBeef_df
```

```python
N_animals_DairyCow = 0
N_animals_DairyHeifer = 0
N_animals_DairyCalf = 0
# N_animals_BeefCow = 0
N_animals_BeefCalf = 0

import time

waste_data          = pd.read_csv('cereslibrary/PTechs/feedstock_data/waste_data_input.csv', sep=",", header=0)
waste_data = waste_data.set_index('Livestock type', drop=False)

tech_selected=[]
process_economic_results = {}
economic_results = {}
results_results = {}
manure = {}

time_start = time.time()
counter = 0
for i in np.arange(0,len(IndianaBeef_df.index)):
    AU = IndianaBeef_df.iloc[i]['Animal Units']
    N_animals_BeefCow = AU*waste_data.loc['Beef Cow','AU_equivalence']
    
    latitude = IndianaBeef_df.iloc[i]['latitude']
    longitude = IndianaBeef_df.iloc[i]['longitude']
    
    results_models = main_function(N_animals_DairyCow, N_animals_DairyHeifer, N_animals_DairyCalf, N_animals_BeefCow, N_animals_BeefCalf, AD_eval, SL_eval, biogas_product, discount_rate, RINs_methane, REC, P_credits, CapitalCost_incentive)
    F_animals_BeefCow = (N_animals_BeefCow/waste_data.loc['Beef Cow','AU_equivalence']*waste_data.loc['Beef Cow','Weight'])/24/3600*24*3600*365
    process_economic_results[IndianaBeef_df.iloc[i]['Address']] = results_models['process_economic_results_summary']
    economic_results[IndianaBeef_df.iloc[i]['Address']] = results_models['economic_results_summary']
    results_results[IndianaBeef_df.iloc[i]['Address']] = results_models['results_summary']
    manure[IndianaBeef_df.iloc[i]['Address']] = F_animals_BeefCow
    
    results_GIS = IndicatorsDatabase(longitude, latitude)

    raw_data = results_models['raw_data']
    element1 = results_GIS['element1']
    element2 = results_GIS['element2']
    element3 = results_GIS['element3']
    
#     print(element2)

    results_MCDA = MCDA_module(raw_data, element1,element2,element3, N, weighting_type, user_weight)

    tech_selected.append(results_MCDA['Rank_weightened_short_df'].iloc[0].name)
    

    counter = counter+1

time_end = time.time()
time_elapsed = time_end - time_start
IndianaBeef_df['tech_selected']=tech_selected
```

```python
manure_list = []
mass_recovered = []
process_capital_cost = []
process_operation_cost = []
net_revenue = []
NPV = []

for i in IndianaBeef_df.index:
    tech = IndianaBeef_df.loc[i,'tech_selected']
    address = IndianaBeef_df.loc[i,'Address']
    manure_i = float(manure[address])
    mass_recovered_i = float(results_results[address].loc[tech,'Mass TP recovered (kg/year)'])
    process_capital_cost_i = float(process_economic_results[address].loc[tech,'Process capital cost (USD)'])
    process_operation_cost_i = float(process_economic_results[address].loc[tech,'Process operation cost (USD/year)'])
    net_revenue_i = float(process_economic_results[address].loc[tech,'Net revenue (USD/year)'])
    NPV_i = float(process_economic_results[address].loc[tech,'NPV (USD)'])
    manure_list.append(manure_i)
    mass_recovered.append(mass_recovered_i)
    process_capital_cost.append(process_capital_cost_i)
    process_operation_cost.append(process_operation_cost_i)
    net_revenue.append(net_revenue_i)
    NPV.append(NPV_i)

# mass_recovered
# # process_capital_cost
# process_operation_cost
# net_revenue
# results_beef_total = results1
IndianaBeef_df['Manure generated (kg/year)'] = manure_list
IndianaBeef_df['Phosphorus released generated (kg/year)'] = IndianaBeef_df['Manure generated (kg/year)']*0.08/100
IndianaBeef_df['Mass TP recovered (kg/year)'] = mass_recovered
IndianaBeef_df['Process capital cost (USD)'] = process_capital_cost
IndianaBeef_df['Process operation cost (USD/year)'] = process_operation_cost
IndianaBeef_df['Net revenue (USD/year)'] = net_revenue
IndianaBeef_df['NPV (USD)'] = NPV
IndianaBeef_df['Profitable'] = np.where(IndianaBeef_df['Net revenue (USD/year)']>=0, 'Yes', 'No')
IndianaBeef_df['Recommended'] = np.where(IndianaBeef_df['NPV (USD)']>=0, 'Yes', 'No')
IndianaBeef_df
```

```python
IndianaBeef_df.to_csv('CAFOs_Indiana/Results/Beef/Indiana_Beef_Results_Pcredits'+str(P_credits)+'_REC'+str(REC)+'_'+biogas_product+'.csv')
time_elapsed
```
